```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load (Remember to Change These)
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

school_name = list(school_data['school_name'])



```

## Local Government Area Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average maths score 

* Calculate the average reading score

* Calculate the percentage of students with a passing maths score (50 or greater)

* Calculate the percentage of students with a passing reading score (50 or greater)

* Calculate the percentage of students who passed maths **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Total School
total_School = len(school_data_complete['School ID'].unique())

# Total student
total_student = len(school_data_complete['Student ID'].unique())

# Total budget
total_budget = school_data['budget'].sum()

# Math avg
math_avg = school_data_complete['maths_score'].mean()

# Reading Avg
read_avg = school_data_complete['reading_score'].mean()

# % of math passers
filt = school_data_complete['maths_score'] >= 50
math_pass_rows_df = school_data_complete.loc[filt]
percent_math_pass=(len(math_pass_rows_df['maths_score'])/total_student)*100

# % of reading passers
filt = school_data_complete['reading_score'] >= 50
reading_pass_rows_df = school_data_complete.loc[filt]
percent_reading_pass=(len(reading_pass_rows_df['reading_score'])/total_student)*100

# % of math & reading passers
filt = ((school_data_complete['reading_score'] >= 50) &  (school_data_complete['maths_score'] >= 50))
reading_math_pass_rows_df = school_data_complete.loc[filt]
percent_reading_math_pass=(len(reading_math_pass_rows_df['reading_score'])/total_student)*100

# create local gov area summary df
data = {"Total Schools":[total_School],"Total Students":[total_student],"Total Budget":[total_budget],
        "Average Maths Score":[math_avg],"Average Reading Score":[read_avg],"% Passing Maths":[percent_math_pass],
        "% Passing Reading":[percent_reading_pass],"% Overall Passing":[percent_reading_math_pass]}

area_summary = pd.DataFrame(data)
area_summary.loc[:, "Total Budget"] ='$'+ area_summary["Total Budget"].map('{:,.2f}'.format)
area_summary.loc[:, "Total Students"] = area_summary["Total Students"].map('{:,}'.format)

area_summary


```

## School Summary

* Create an overview table that summarises key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed maths **and** reading.)
  
* Create a dataframe to hold the above results

```python
group_by_school = school_data_complete.groupby(['school_name'])

# Budget per student
school_data['Per Student Budget'] = school_data['budget']/school_data['size']

# Math average score by school
math_score_Average_bySchool_df = pd.DataFrame(group_by_school['maths_score'].mean())

# Reading average score by school
reading_Score_Ave_byShcool_df = pd.DataFrame(group_by_school['reading_score'].mean())

# % of math passers
accu = {'school_name':[],'math_passer_percent':[]}
for i in school_name:
        data = school_data_complete.groupby('school_name')[['maths_score','size']].get_group(i)
        filt = (data['maths_score'] >= 50)
        stu_count = int(data['size'].mean())
        data = data.loc[filt]
        overall_passer = ((data['maths_score'].count())/stu_count)*100
        accu['school_name'].append(i)
        accu['math_passer_percent'].append(overall_passer)
per_math_passer = pd.DataFrame(accu)

# % of reading passers
accu = {'school_name':[],'reading_passer_percent':[]}
for i in school_name:
        data = school_data_complete.groupby('school_name')[['reading_score','size']].get_group(i)
        filt = (data['reading_score'] >= 50)
        stu_count = int(data['size'].mean())
        data = data.loc[filt]
        overall_passer = ((data['reading_score'].count())/stu_count)*100
        accu['school_name'].append(i)
        accu['reading_passer_percent'].append(overall_passer)
per_reading_passer = pd.DataFrame(accu)

# % of overall passers
accu = {'school_name':[],'overall_passing_percent':[]}
for i in school_name:
        data = school_data_complete.groupby('school_name')[['maths_score','reading_score','size']].get_group(i)
        filt = ((data['reading_score'] >= 50) &  (data['maths_score'] >= 50))
        stu_count = int(data['size'].mean())
        data = data.loc[filt]
        overall_passer = ((data['reading_score'].count())/stu_count)*100
        accu['school_name'].append(i)
        accu['overall_passing_percent'].append(overall_passer)
        
overall_passers = pd.DataFrame(accu)


merged_df = pd.merge(pd.merge(math_score_Average_bySchool_df,reading_Score_Ave_byShcool_df, on="school_name",how='left'),per_math_passer,on="school_name",how='left')
merged_df = pd.merge(pd.merge(merged_df,per_reading_passer,on="school_name",how='left'),overall_passers,on="school_name",how='left')
merged_df = pd.merge(merged_df,school_data,on="school_name",how="left")

per_school_summary = merged_df.copy(deep=True)

per_school_summary.drop(columns=['School ID'],inplace=True)

per_school_summary.rename(columns={'type':'School Type',
                         'size':'Total Students',
                         'budget':'Total School Budget',
                         'maths_score':'Average Maths Score',
                         'reading_score':'Average Reading Score',
                         'math_passer_percent':'% Passing Maths',
                         'reading_passer_percent':'% Passing Reading',
                         'overall_passing_percent':'% Overall Passing'
                          },inplace=True)


per_school_summary.loc[:, "Total School Budget"] ='$'+ per_school_summary["Total School Budget"].map('{:,.2f}'.format)
per_school_summary.loc[:, "Per Student Budget"] = '$'+ per_school_summary["Per Student Budget"].map('{:,.2f}'.format)
new_cols = ["school_name","School Type","Total Students","Total School Budget","Per Student Budget","Average Maths Score",
            "Average Reading Score","% Passing Maths","% Passing Reading","% Overall Passing"]
per_school_summary=per_school_summary.reindex(columns=new_cols)
per_school_summary.set_index('school_name',inplace=True)
per_school_summary

```

## Top Performing Schools (By % Overall Passing)

* Sort and display the top five performing schools by % overall passing.

```python
top_schools = per_school_summary.sort_values(by=['% Overall Passing'],ascending=False)
top_schools.head(5)
```

## Maths Scores by Year

* Create a table that lists the average maths score for students of each year level (9, 10, 11, 12) at each school.

  * Create a pandas series for each year. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
years = list(school_data_complete['year'].unique())

# Average maths & Reading score for sutdents of each year level (9, 10, 11, 12) at each school.

accu = {'school_name':[]}
for i in years:
        year_math = "Year " + str(i) + " math"
        year_reading = "Year " + str(i) + " reading"
        data = school_data_complete.groupby('year')[['maths_score','reading_score','school_name','year']].get_group(i)
        accu[year_math] = []
        accu[year_reading] = []
        for j in school_name:
            data_byYear = data.groupby('school_name')[['maths_score','reading_score','year','school_name']].get_group(j)
            avg_math = data_byYear['maths_score'].mean()
            avg_reading = data_byYear['reading_score'].mean()
            if j not in accu['school_name']:
                accu['school_name'].append(j)
                accu[year_math].append(avg_math)
                accu[year_reading].append(avg_reading)
            else:
                accu[year_math].append(avg_math)
                accu[year_reading].append(avg_reading)
            
            
maths_scores_by_year = pd.DataFrame({"school_name":accu['school_name'],"Year 9":accu['Year 9 math'],
                         "Year 10":accu['Year 10 math'],"Year 11":accu['Year 11 math'],
                         "Year 12":accu['Year 12 math']})

# Average maths score for sutdents of each year level (9, 10, 11, 12) at each school.

maths_scores_by_year.set_index('school_name',inplace=True)
maths_scores_by_year
     
        
        
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Create bins in which to place values based upon spending per student
bins = [0,585,630,645,680]

# Create labels for these bins
group_labels = ["<$584", "$585-630", "$630-645", "$645-680"]
spending_summary = merged_df.copy(deep=True)
spending_summary['Spending Ranges (Per Student)'] = pd.cut(spending_summary['Per Student Budget'],
                                                           bins,labels=group_labels)
spending_summary.drop(columns=['School ID','type','budget','size','school_name','Per Student Budget'],inplace=True)
spending_summary.rename(columns={
                         'maths_score':'Average Maths Score',
                         'reading_score':'Average Reading Score',
                         'math_passer_percent':'% Passing Maths',
                         'reading_passer_percent':'% Passing Reading',
                         'overall_passing_percent':'% Overall Passing'
                          },inplace=True)
#score_bySchool_spending.loc[:, "Average Maths Score"] =score_bySchool_spending["Average Maths Score"].map('{:,.2f}'.format)
spending_summary.set_index('Spending Ranges (Per Student)',inplace=True)
spending_summary.groupby('Spending Ranges (Per Student)').mean()
```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
# Create bins in which to place values based upon school size
bins = [0,1000,2000,5000]

# Create labels for these bins
group_labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]

size_summary = merged_df.copy(deep=True)
size_summary['School Size'] = pd.cut(size_summary['size'],bins,labels=group_labels)
size_summary.drop(columns=['School ID','type','budget','size','school_name','Per Student Budget'],inplace=True)
size_summary.rename(columns={
                         'maths_score':'Average Maths Score',
                         'reading_score':'Average Reading Score',
                         'math_passer_percent':'% Passing Maths',
                         'reading_passer_percent':'% Passing Reading',
                         'overall_passing_percent':'% Overall Passing'
                          },inplace=True)
size_summary.loc[:, "Average Maths Score"] =size_summary["Average Maths Score"].map('{:,.2f}'.format)

size_summary.set_index('School Size',inplace=True)

size_summary.groupby('School Size').mean()

```
